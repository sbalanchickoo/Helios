```python
df3 = pd.DataFrame(
    [
        ['John', 7, 8]
        , ['Jill', 6, 6]
        , ['Jack', 5, 4]
        , ['Joan', 5, 6]
    ]
    , columns=['student', 'test1 score', 'test2 score']
    , index=['s1','s2','s3', 's4']
)
```

```python
# Return where test score did not improve
df3['test2 score'] <= df3['test1 score']
```

```python
# Filtering using boolean expression, to return indices
df3[df3['test2 score'] <= df3['test1 score']]
```

```python
# Indices of rows with any null values
df.isnull().any(axis=1).any()
```

```python
null_rows = pd.Series(df.notnull().any(axis=1))
df['date'][null_rows].head(5)
```
